```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Bar plots - regular, stacked, grouped


labels = ('Python', 'C++', 'Java', 'Perl', 'C#')
num_users = [12,8,15,4,6]
index = np.arange(len(labels))                   # Converting categorical languages to numerical values

plt.bar(index, num_users, align='center', alpha=0.5)
plt.xticks(index, labels)                       # Representing index values back as the respective lang label
plt.xlabel('Language')
plt.ylabel('Num of Users')
plt.title('Programming language user base')
```

```python
# Box Plot - Providing statistical information all the way

data_group1 = np.random.normal(100,10,200)
data_group2 = np.random.normal(80,30,200)
data_group3 = np.random.normal(90,20,200)

data_to_plot = [data_group1, data_group2, data_group3]

plt.boxplot(data_to_plot, patch_artist = True)

plt.xlabel('Group')
plt.ylabel('Scores')
plt.title('Some Group Scores')
```
