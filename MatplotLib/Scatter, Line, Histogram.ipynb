```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Scatter plots - The below scatterplot represents 4 different dimensions of data - x,y,color and size

rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)


plt.scatter(x, y, c=colors, s=sizes, alpha=0.2, cmap='gist_earth')
plt.colorbar()                    # Viewing the colorbar next to the plot


#alpha - Transparency of points
#cmpa - colormaps for creating visually appealing plots. - plt.cm.Blues, plt.cm.autum, gist-earth, viridis
#c - mapped to color scale and s - is in pixels
```

```python
# Line plots


year = [1970, 1980, 1990, 2000, 2010, 2020]
pop_A = [4.9, 5.3, 7.1, 10.7, 13.5, 15.6]
pop_B = [44.4, 55.6, 69.7, 87.1, 95.4, 100.5]

plt.plot(year, pop_A, color='g', label='A')
plt.plot(year, pop_B, color='r', label='B')

leg = plt.legend(loc='best', ncol=1, shadow=True, fancybox=True)
leg.get_frame().set_alpha(0.5)

plt.xlabel('Years')
plt.ylabel('Population in millions')
plt.title('Population over the years')
```

```python
# Histograms

data = np.random.randn(1000)
plt.hist(x=data, bins=30, alpha=0.5, histtype='stepfilled', color='green')
```
