```python
import pandas as pd
```

```python
# Selection of data

movies_df = pd.read_csv('imdb_data.csv', index_col='Title')
movies_df.head(5)
```

```python
# Column as a series

col_series = movies_df['Genre']
print(type(col_series))
col_series.head(5)
```

```python
# Column as a dataframe

col_df = movies_df[['Genre', 'Rating', 'Year']]  # we are passing the columns as a list to get a Dataframe output
print(type(col_df))
col_df.head(5)
```

```python
# Slicing of data - row wise (loc, iloc, ix) - [start:stop:step]
 
movies_df.loc['Guardians of the Galaxy']   # Explicit index based 
```

```python
movies_df.iloc[0]  # Implicit index based - Numerical index
```

```python
# Slicing with Multiple Rows

movies_df.loc['Guardians of the Galaxy' : 'Sing']
```

```python
movies_df.iloc[0:4]
```

```python
# Slicing both rows and columns

movies_df.loc[:'Sing', :'Director']
```

```python
movies_df.iloc[:4, :4]
```

```python
# Using ix - hybrid slicing

movies_df.ix[:4, :'Director'] # ix is deprecated since pandas 0.20.0
```

```python
# Conditional data selection and filtering - because we never know what indexes we want to work on! (implicit/explicit)

movies_df[movies_df['Year'] == 2016] # Movies released in 2016
```

```python
movies_df[movies_df['Rating'] > 8.0] # movies with rating greater than 8.0
```

```python
movies_df[(2016 >= movies_df['Year'] >= 2010)] 

# The following error is thrown because the first condition evaluates to True and True is compared >= 2010
```

```python
movies_df[((2016 >= movies_df['Year']) & (movies_df['Year'] >= 2010)) 
          & (movies_df['Rating'] < 6.0) 
          & (movies_df['Revenue (Millions)'] > movies_df['Revenue (Millions)'].quantile(0.75))
         ]
```
