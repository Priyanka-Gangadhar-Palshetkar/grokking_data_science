```python
# Grouping 

import pandas as pd

movies_df = pd.read_csv('imdb_data.csv')
movies_df.head()
```

```python
# Revenue each director earned and sum of ratings

movies_df.groupby('Director')[['Revenue (Millions)', 'Rating']].sum()
```

```python
# Average rating for each director

movies_df.groupby('Director')[['Rating']].mean()
```

```python
# Sorting - revenue per director sorted by highest revenues

movies_df.groupby('Director')[['Revenue (Millions)']].sum().sort_values(['Revenue (Millions)'], ascending=False)
```

```python
# Sorting - top 10 movies which had both the highest revenue and rating

movies_df.sort_values(['Revenue (Millions)', 'Rating'], ascending=False).head(10)
```
