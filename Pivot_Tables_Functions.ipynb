```python
import pandas as pd

movies_df = pd.read_csv('imdb_data.csv', index_col="Title")
```

```python
# Creating a pivot table - Annual trend in revenue per Director 

movies_df.pivot_table('Revenue (Millions)', index='Director', aggfunc='sum', columns='Year').head(5)
```

```python
# Applying functions on our dataframes - Vectorization is in play!

def rating_bucket(x):
    if x >= 8.0:
        return "Great"
    elif x >= 7.0:
        return "Good"
    elif x >= 6.0:
        return "Average"
    else:
        return "Bad"

movies_df['Rating Category'] = movies_df['Rating'].apply(rating_bucket)
movies_df.head(5)[['Rating', 'Rating Category']]
```
